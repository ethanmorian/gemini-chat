```python
import pandas as pd

df = pd.read_csv('senario.csv')
df
```

```python
dataset = []

for _, row in df.iterrows():
    input_text = row["keywords"]
    output_text = row["category"]
    # 데이터셋에 추가
    dataset.append({
        "input": input_text,
        "output": output_text,
    })

for item in dataset:
    print("Input:")
    print(item["input"])
    print("Output:")
    print(item["output"])
```

```python
df = pd.DataFrame(dataset)

jsonl_path = "dataset.jsonl"
df.to_json(jsonl_path, orient="records", lines=True, force_ascii=False)
```
